### <font color='DarkOrange'>**Задание 9 [0.5 баллa]**</font>

Заполните пропуски пустой строкой для текстовых колонок, и удалите строки с пропусками для колонок с категориальными значениями.

<font color='LightSteelBlue'>**Подсказка**</font>

Значения <font color='CornflowerBlue'>**категориальной**</font> переменной определяют факт принадлежности к какой-то категории, категорий при этом должно быть ограниченное кол-во сильно меньшее размера датасета.

<font color='CornflowerBlue'>**Категориальная переменная**</font> - колонка, которая имеет ограниченное число уникальных значений. Вообще говоря, может быть как типа object (текстовые), так и int (численные). В случае int можно считать, что категории закодированы числами. В таком случае значения нельзя сравнивать между собой и совершать над ними арифметические операции. В этом задании категориальные переменные только типа object.

<font color='OrangeRed'>**Внимание!**</font> Для проверки корректности выведите кол-во пропусков во всех столбцах и общее кол-во строк в таблице - должна быть 3021 запись. Ошибка при удалении строк может привести к неправильному выполнению следующих заданий.




```python
df_cleaned = df.copy()

categorical_cols = []
text_cols = []

for col in df_cleaned.columns:
    if df_cleaned[col].dtype == 'object':
        n_unique = df_cleaned[col].nunique()
        if n_unique < 50 or n_unique < len(df_cleaned) * 0.05:
            categorical_cols.append(col)
        else:
            text_cols.append(col)

for col in text_cols:
    if df_cleaned[col].isnull().any():
        df_cleaned[col].fillna('', inplace=True)

rows_before = len(df_cleaned)
df_cleaned = df_cleaned.dropna(subset=categorical_cols)
rows_after = len(df_cleaned)

print(f"\nУдалено строк с пропусками в категориальных колонках: {rows_before - rows_after}")
print(f"Осталось строк: {rows_after}")

df = df_cleaned
```

<font color='MediumOrchid'>**Ваш ответ тут:**</font>

Удалено строк с пропусками в категориальных колонках: 0
Осталось строк: 3021

- <font color='DarkSeaGreen'>**bar plot**</font> — столбчатая диаграмма, показывающий взаимосвязь количественной величины от категориальной.

#### <font color='DarkOrange'>**Задача 1 [0.3 + 0.7 баллов]**</font>

Вам дан двумерный непустой массив X размера n x m вещественных чисел.
Создайте на его основе массив состоящий: из каждого 4го (начиная с 0го и до конца) элемента по оси размерности n (axis=0), начиная с 120 до 500 (не включитально) с шагом 5 элементов по оси размерности m (axis=1).

Гарантируется, что m >= 500

```python
def get_part_of_array(X: List[List[float]]) -> List[List[float]]:
    """
    X - двумерный массив вещественных чисел размера n x m. Гарантируется что m >= 500
    Вернуть: двумерный массив, состоящий из каждого 4го элемента по оси размерности n
    и c 120 по 500 c шагом 5 по оси размерности m
    """
    result = []
    for i in range(0, len(X), 4):
        row = []
        for j in range(120, 500, 5):
            row.append(X[i][j])
        result.append(row)
    return result
```

```python
def get_part_of_array(X: np.ndarray) -> np.ndarray:
    """
    X - двумерный массив размера n x m. Гарантируется что m >= 500
    Вернуть: двумерный массив, состоящий из каждого 4го элемента по оси размерности n
    и c 120 по 500 c шагом 5 по оси размерности m
    """
    return X[::4, 120:500:5]
```

#### <font color='DarkOrange'>**Задача 2 [0.3 + 0.7 баллов]**</font>

Вам дан двумерный непустой массив X целых чисел размера n x m.

Посчитайте сумму неотрицательных элементов на главной диагонали прямоугольной матрицы X. Если неотрицательных элементов на диагонали нет, то верните -1.

  **Пример:**
  
      Input: X = np.array([[-1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
      Output: 3

```python
def sum_non_neg_diag(X: List[List[int]]) -> int:
    """
    Вернуть  сумму неотрицательных элементов на диагонали прямоугольной матрицы X.
    Если неотрицательных элементов на диагонали нет, то вернуть -1
    """
    total = 0
    has_non_neg = False
    n = len(X)
    m = len(X[0]) if n > 0 else 0
    diag_len = min(n, m)

    for i in range(diag_len):
        if X[i][i] >= 0:
            total += X[i][i]
            has_non_neg = True

    return total if has_non_neg else -1
```

```python
def sum_non_neg_diag(X: np.ndarray) -> int:
    """
    Вернуть  сумму неотрицательных элементов на диагонали прямоугольной матрицы X.
    Если неотрицательных элементов на диагонали нет, то вернуть -1
    """
    diag = np.diag(X)
    non_neg = diag[diag >= 0]
    return int(non_neg.sum()) if non_neg.size > 0 else -1
```

#### <font color='DarkOrange'>**Задача 3 [0.3 + 0.7 баллов]**</font>

Дан двумерный непустой массив X вещественных чисел размера n x m. Считая для каждого столбца отдельно, замените в копии массива те значения, которые больше 1.5 средних по столбцу или меньше 1/4 среднего по столбцу на число -1

<font color='LightSteelBlue'>**Подсказка**</font>

Возможно пригодится функция deepcopy из модуля copy

```python
def replace_values(X: List[List[float]]) -> List[List[float]]:
    """
    X - двумерный массив вещественных чисел размера n x m.
    По каждому столбцу нужно почитать среднее значение M.
    В каждом столбце отдельно заменить: значения, которые < 0.25M или > 1.5M на -1
    Вернуть: двумерный массив, копию от X, с измененными значениями по правилу выше
    """
    result = deepcopy(X)
    n = len(X)
    m = len(X[0]) if n > 0 else 0

    for j in range(m):
        # Calculate column mean
        col_sum = 0.0
        for i in range(n):
            col_sum += X[i][j]
        mean = col_sum / n

        # Replace values that are < 0.25 * mean or > 1.5 * mean
        for i in range(n):
            if result[i][j] < 0.25 * mean or result[i][j] > 1.5 * mean:
                result[i][j] = -1

    return result
```

```python
def replace_values(X: np.ndarray) -> np.ndarray:
    """
    X - двумерный массив вещественных чисел размера n x m.
    По каждому столбцу нужно почитать среднее значение M.
    В каждом столбце отдельно заменить: значения, которые < 0.25M или > 1.5M на -1
    Вернуть: двумерный массив, копию от X, с измененными значениями по правилу выше
    """
    result = X.copy()
    col_means = np.mean(X, axis=0)
    mask = (result < 0.25 * col_means) | (result > 1.5 * col_means)
    result[mask] = -1
    return result
```
