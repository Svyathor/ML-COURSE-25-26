```python
df_cleaned = df.copy()

categorical_cols = []
text_cols = []

for col in df_cleaned.columns:
    if df_cleaned[col].dtype == 'object':
        n_unique = df_cleaned[col].nunique()
        if n_unique < 50 or n_unique < len(df_cleaned) * 0.05:
            categorical_cols.append(col)
        else:
            text_cols.append(col)

for col in text_cols:
    if df_cleaned[col].isnull().any():
        df_cleaned[col].fillna('', inplace=True)

rows_before = len(df_cleaned)
df_cleaned = df_cleaned.dropna(subset=categorical_cols)
rows_after = len(df_cleaned)

print(f"\n      : {rows_before - rows_after}")
print(f" : {rows_after}")

df = df_cleaned
```

#### <font color='DarkOrange'>** 1 [0.3 + 0.7 ]**</font>

     X  n x m  .
     :   4 (  0   )     n (axis=0),   120  500 ( )   5     m (axis=1).

,  m >= 500

```python
def get_part_of_array(X: List[List[float]]) -> List[List[float]]:
    """
    X -      n x m.   m >= 500
    :  ,    4     n
     c 120  500 c  5    m
    """
    result = []
    for i in range(0, len(X), 4):
        row = []
        for j in range(120, 500, 5):
            row.append(X[i][j])
        result.append(row)
    return result
```

```python
def get_part_of_array(X: np.ndarray) -> np.ndarray:
    """
    X -    n x m.   m >= 500
    :  ,    4     n
     c 120  500 c  5    m
    """
    return X[::4, 120:500:5]
```

 **:**

      Input: X = [[1, 2, 3, ..., 500], [1, 2, 3, ..., 500], [1, 2, 3, ..., 500], [1, 2, 3, ..., 500], [1, 2, 3, ..., 500]]
      Output: [[121, 126, 131, ..., 496]]

#### <font color='DarkOrange'>** 2 [0.3 + 0.7 ]**</font>

     X    n x m.

         X.      ,   -1.

  **:**
  
      Input: X = np.array([[-1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
      Output: 3

```python
def sum_non_neg_diag(X: List[List[int]]) -> int:
    """
             X.
         ,   -1
    """
    total = 0
    has_non_neg = False
    n = len(X)
    m = len(X[0]) if n > 0 else 0
    diag_len = min(n, m)

    for i in range(diag_len):
        if X[i][i] >= 0:
            total += X[i][i]
            has_non_neg = True

    return total if has_non_neg else -1
```

```python
def sum_non_neg_diag(X: np.ndarray) -> int:
    """
             X.
         ,   -1
    """
    diag = np.diag(X)
    non_neg = diag[diag >= 0]
    return int(non_neg.sum()) if non_neg.size > 0 else -1
```

#### <font color='DarkOrange'>** 3 [0.3 + 0.7 ]**</font>

    X    n x m.     ,      ,   1.5      1/4      -1

<font color='LightSteelBlue'>****</font>

   deepcopy   copy

```python
def replace_values(X: List[List[float]]) -> List[List[float]]:
    """
    X -      n x m.
           M.
        : ,  < 0.25M  > 1.5M  -1
    :  ,   X,      
    """
    result = deepcopy(X)
    n = len(X)
    m = len(X[0]) if n > 0 else 0

    for j in range(m):
        # Calculate column mean
        col_sum = 0.0
        for i in range(n):
            col_sum += X[i][j]
        mean = col_sum / n

        # Replace values that are < 0.25 * mean or > 1.5 * mean
        for i in range(n):
            if result[i][j] < 0.25 * mean or result[i][j] > 1.5 * mean:
                result[i][j] = -1

    return result
```

```python
def replace_values(X: np.ndarray) -> np.ndarray:
    """
    X -      n x m.
           M.
        : ,  < 0.25M  > 1.5M  -1
    :  ,   X,      
    """
    result = X.copy()
    col_means = np.mean(X, axis=0)
    mask = (result < 0.25 * col_means) | (result > 1.5 * col_means)
    result[mask] = -1
    return result
```
